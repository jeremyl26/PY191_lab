```python
import pandas as pd

data = pd.read_csv("aapl.csv")

prices = data['Close'].tolist()

sma_20 = []
sma_50 = []

for i in range(len(prices)):
    if i < 19:
        sma_20.append(None)  
    else:
        total_20 = 0
        for j in range(i - 19, i + 1):
            total_20 += prices[j]
        sma_20.append(total_20 / 20)

    if i < 49:
        sma_50.append(None)  
    else:
        total_50 = 0
        for j in range(i - 49, i + 1):
            total_50 += prices[j]
        sma_50.append(total_50 / 50)

capital = 100.00      
shares_held = 0.0     
position = "cash"     

for i in range(50, len(prices)):
    current_price = prices[i]
    yesterday_short = sma_20[i - 1]
    yesterday_long = sma_50[i - 1]

    if position == "cash" and yesterday_short > yesterday_long:
        shares_held = capital / current_price
        capital = 0.0
        position = "stock"

    elif position == "stock" and yesterday_short < yesterday_long:
        capital = shares_held * current_price
        shares_held = 0.0
        position = "cash"

if position == "stock":
    final_value = shares_held * prices[-1]
else:
    final_value = capital

bh_shares = 100.00 / prices[0]
final_bh_value = bh_shares * prices[-1]

print(f"--- 5-Year Results ---")
print(f"Starting: $100.00")
print(f"Strategy Final Value: ${final_value:.2f}")
print(f"Hold Final Value: ${final_bh_value:.2f}")
```
